Comptador i endreçador de paraules d'un text.

#### Instrucciones de Uso

#### Habilitar y Procesar Texto de un Archivo

1. **Función `habilitar_texto(txt_cliente)`**:
   - Esta función toma como entrada la ruta de un archivo de texto (`txt_cliente`) y devuelve el contenido completo del archivo como una cadena de texto.
   - Utiliza codificación UTF-8 para leer el archivo de texto.

2. **Función `contar_palabras(texto)`**:
   - Define el alfabeto en minúsculas (`abcdefghijklmnopqrstuvwxyz`) para organizar las palabras según su letra inicial.
   - Elimina los signos de puntuación del texto utilizando una lista de signos predefinida.
   - Convierte todas las palabras del texto a minúsculas y las divide en una lista.
   - Cuenta la frecuencia de cada palabra y las agrupa según su letra inicial en un diccionario (`diccionario_paraules`).
   - Ordena alfabéticamente las palabras dentro de cada grupo de letras y devuelve un diccionario ordenado (`diccionario_ordenat`).

#### Ejecución del Código

- Se define el nombre del archivo de texto (`txt_cliente`) que se va a procesar.
- Se utiliza la función `habilitar_texto` para cargar el contenido del archivo especificado.
- Se llama a la función `contar_palabras` con el texto cargado para contar y ordenar las palabras por letra inicial.

#### Imprimir Resultados

- Después de contar y ordenar las palabras:
  - Se imprime cada letra del alfabeto junto con las palabras que comienzan con esa letra y su frecuencia.
  - Las palabras están listadas en orden alfabético bajo cada letra.

#### Nota
- Asegúrate de que el archivo `tu_me_quieres_blanca.txt` exista en la misma ubicación desde donde se ejecuta el script.
- El archivo debe contener el texto del cual se desean contar y ordenar las palabras según la letra inicial.
- Asegúrate de tener los permisos necesarios para acceder y leer el archivo.


```python
# Función para habilitar el contenido del archivo
def habilitar_texto(txt_cliente):
    with open(txt_cliente, "r", encoding="utf-8") as archivo:
        return archivo.read()

# Función para contar y ordenar las palabras
def contar_palabras(texto):
    # Definición del alfabeto 
    alfabeto = 'abcdefghijklmnopqrstuvwxyz'
    
    # Definir un diccionario vacío para almacenar las palabras por letra inicial
    diccionario_paraules = {letra: {} for letra in alfabeto}
    
    # Eliminar signos de puntuación y dividir el texto en palabras
    signos_puntuacion = '''!()-[]{};:'",<>./?@#$%^&*_~'''
    for signo in signos_puntuacion:
        texto = texto.replace(signo, '')
    
    palabras = sorted(texto.lower().split())
    
    # Contar la frecuencia de cada palabra y agruparlas por letra inicial
    for palabra in palabras:
        primera_letra = palabra[0]
        if primera_letra in diccionario_paraules:
            if palabra in diccionario_paraules[primera_letra]:
                diccionario_paraules[primera_letra][palabra] += 1
            else:
                diccionario_paraules[primera_letra][palabra] = 1
    
    # Ordenar las palabras dentro de cada letra inicial
    diccionario_ordenat = {}
    for letra, paraules_dict in diccionario_paraules.items():
        if paraules_dict:
            diccionario_ordenat[letra] = dict(sorted(paraules_dict.items()))
    
    return diccionario_ordenat

# Nombre del archivo cliente
txt_cliente = "tu_me_quieres_blanca.txt"

# Habilitar el texto del archivo
texto = habilitar_texto(txt_cliente)

# Contar y ordenar las palabras
diccionario_ordenat = contar_palabras(texto)

# Imprimir resultados de manera bonita
for letra, palabras_dict in diccionario_ordenat.items():
    print(f"Lletra {letra.upper()}:")
    for palabra, conteo in palabras_dict.items():
        print(f"  {palabra}: {conteo}")
    print()  # Para separar las secciones por letra
```

```python
# Funció per llegir el fitxer de text i retornar el contingut
def llegir_fitxer(nom_fitxer):
    with open(nom_fitxer, "r", encoding="utf-8") as archivo:
        return archivo.read()

# Funció per contar les paraules i ordenar-les segons les entrades habituals del diccionari
def contar_paraules(text):
    # Definim els signes de puntuació que volem eliminar
    signes_puntuacio = '''!()-[]{};:'",<>./?@#$%^&*_~'''
    
    # Eliminem els signes de puntuació del text
    for signe in signes_puntuacio:
        text = text.replace(signe, '')
    
    # Convertim el text a minúscules i el separem en paraules
    paraules = text.lower().split()
    
    # Inicialitzem el diccionari amb les lletres de l'A a la Z
    diccionari_paraules = {chr(lletra): {} for lletra in range(ord('a'), ord('z') + 1)}
    
    # Contem la freqüència de cada paraula
    for paraula in paraules:
        primera_lletra = paraula[0]
        if primera_lletra in diccionari_paraules:
            if paraula in diccionari_paraules[primera_lletra]:
                diccionari_paraules[primera_lletra][paraula] += 1
            else:
                diccionari_paraules[primera_lletra][paraula] = 1
    
    # Ordenem les paraules dins de cada lletra
    diccionari_ordenat = {}
    for lletra, paraules_dict in diccionari_paraules.items():
        if paraules_dict:
            diccionari_ordenat[lletra] = dict(sorted(paraules_dict.items()))
    
    return diccionari_ordenat

# Llegim el fitxer
nom_fitxer = "tu_me_quieres_blanca.txt"
text = llegir_fitxer(nom_fitxer)

# Comptem i ordenem les paraules
diccionari_ordenat = contar_paraules(text)

# Imprimim els resultats de manera bonica
for lletra, paraules_dict in diccionari_ordenat.items():
    print(f"Lletra {lletra.upper()}:")
    for paraula, conteo in paraules_dict.items():
        print(f"  {paraula}: {conteo}")
    print()  # Per separar les seccions per lletra

```
