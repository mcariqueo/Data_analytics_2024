### Exercici 1: Calculadora de l'índex de massa corporal
Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. La funció ha de classificar el resultat en les seves respectives categories

----

##### CALCULADORA DEL ÍNDICE DE MASA CORPORAL (IMC)

##### INSTRUCCIONES

1. **Ingrese su peso corporal:**
   - Introduzca su peso corporal en kilogramos (kg).
   - El peso debe estar entre 40 y 150 kg. Si el valor ingresado no está en este rango, se le pedirá que lo ingrese nuevamente.

2. **Ingrese su altura:**
   - Introduzca su altura en centímetros (cm).
   - La altura debe estar entre 100 y 250 cm. Si el valor ingresado no está en este rango, se le pedirá que lo ingrese nuevamente.

3. **Calcule su IMC:**
   - Una vez ingresados correctamente su peso y altura, se calculará su Índice de Masa Corporal (IMC) utilizando la fórmula: `IMC = peso / (altura en metros al cuadrado)`.
   - El resultado se redondeará a dos decimales.

4. **Obtenga la clasificación de su IMC:**
   - Según su IMC calculado, se le proporcionará una clasificación:
     - Bajo peso: IMC < 18.5
     - Peso normal: 18.5 ≤ IMC < 25
     - Sobrepeso: 25 ≤ IMC < 30
     - Obesidad: IMC ≥ 30

5. **Considere las limitaciones:**
   - Tenga presente que la clasificación del IMC tiene limitaciones y cualquier interpretación debe ser acompañada por un profesional de la salud.


````python
# FUNCION DE CALULAR IMC # CLASIFICACIÓN DEL IMC

def calcular_imc_clasificacion(peso_corporal, altura_cm):
    altura_m = altura_cm / 100                              # Convierto la altura de centímetros a metros
    imc      = peso_corporal / (altura_m ** 2)              # Calculo el IMC utilizando la fórmula: peso / (altura al cuadrado)
    imc      = round(imc, 2)                                # Redondeo el resultado a dos decimales
    
    if   imc  <  18.5     : clase = "Bajo peso"
    elif 18.5 <= imc < 25 : clase = "Peso normal"
    elif 25   <= imc < 30 : clase = "Sobrepeso"
    else                  : clase = "Obesidad"
    return imc , clase

# BUCLE PARA SOLICITAR INFORMACION AL USUARIO Y VALIDARLA
while True:
    try:
        peso_corporal = float(input("Ingrese su peso corporal en Kg: "))
        if not 40 <= peso_corporal <= 150:
            raise ValueError("El peso corporal debe estar entre 40 y 150 kg.")
        
        altura_cm = float(input("Ingrese su altura en cm: "))
        if not 100 <= altura_cm <= 250:
            raise ValueError("La altura debe estar entre 100 y 250 cm.")
        
        # Si ambos valores están dentro del rango, salir del bucle externo
        break
    except ValueError as e:
        print(e)
        print("Por favor, intente de nuevo.")

# LLAMO A LAS FUNCIONES DEFINIDAS AL PRINCIPIO DE ESTE CODIGO:
imc, clase   = calcular_imc_clasificacion(peso_corporal, altura_cm)

print(" Calculadora de l'índex de massa corporal (IMC)\n" +
      f"        Su peso es de {peso_corporal} kg y su altura es de {altura_cm / 100} m.\n" +
      f"        Su IMC es {imc}. Clasificación: {clase}\n" +
      "Tenga presente que esta clasificación tiene limitaciones y cualquier interpretación debe ser acompañada por un profesional")
````

### Exercici 2

Convertidor de temperatures.

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona'n almenys 2 i construeix el convertidor.

----

#####  CONVERTIDOR DE TEMPERATURAS

#####  INSTRUCCIONES

1. **Ingrese el valor numérico de la temperatura:**
   - Introduzca el valor de la temperatura que desea convertir.
   
2. **Seleccione la unidad de la temperatura ingresada:**
   - Elija la unidad en la que se encuentra el valor ingresado. Las opciones son:
     1. Celsius (C)
     2. Fahrenheit (F)
     3. Kelvin (K)
     4. Rankine (R)
     5. Réaumur (Re)
   
3. **Seleccione la primera unidad a la que desea convertir:**
   - Elija la primera unidad a la que desea convertir la temperatura ingresada. Las opciones son:
     1. Celsius (C)
     2. Fahrenheit (F)
     3. Kelvin (K)
     4. Rankine (R)
     5. Réaumur (Re)
   
4. **Seleccione la segunda unidad a la que desea convertir:**
   - Elija la segunda unidad a la que desea convertir la temperatura ingresada. Las opciones son:
     1. Celsius (C)
     2. Fahrenheit (F)
     3. Kelvin (K)
     4. Rankine (R)
     5. Réaumur (Re)
----

````python
# FUNCION PARA ELEGIR LAS UNIDADES DE TEMPERATURA
def seleccionar_unidad(mensaje):
    while True:
        opcion = input(mensaje)
        if opcion == "1": return "Celsius"
        elif opcion == "2": return "Fahrenheit"
        elif opcion == "3": return "Kelvin"
        elif opcion == "4": return "Rankine"
        elif opcion == "5": return "Reaumur"
        else:   
            print("Error: Ingrese un número válido.")

# FUNCION PARA INGRESAR EL VALOR NUMERICO DE TEMPERATURA
def ingresar_grados():
    while True:
        try:
            grados = float(input("Ingrese la temperatura: ")) # Como el input es string hago el cast a float()
            return grados
        except ValueError:
            print("Error: Ingrese un valor numérico.")

# FUNCIONES PARA CALCULAR LAS CONVERSIONES: He agreagado todas las conversiones.
def celsius_a_otras(temp):
    return {
        "Fahrenheit"    : (temp * 1.8) + 32, 
        "Kelvin"        :  temp + 273.15,
        "Rankine"       :  temp * 9/5 + 491.67,
        "Reaumur"       :  temp * 4/5
    }

def fahrenheit_a_otras(temp):
    return {
        "Celsius"       : (temp - 32) / 1.8,
        "Kelvin"        : (temp - 32) / 1.8 + 273.15,
        "Rankine"       :  temp + 459.67,
        "Reaumur"       : (temp - 32) * 4/9
    }

def kelvin_a_otras(temp):
    return {
        "Celsius"       : temp - 273.15,
        "Fahrenheit"    : (temp - 273.15) * 1.8 + 32,
        "Rankine"       : temp * 1.8,
        "Reaumur"       : (temp - 273.15) * 4/5
    }

def rankine_a_otras(temp):
    return {
        "Celsius"       : (temp - 491.67) * 5/9,
        "Fahrenheit"    : temp - 459.67,
        "Kelvin"        : temp / 1.8,
        "Reaumur"       : (temp - 491.67) * 4/9
    }

def reaumur_a_otras(temp):
    return {
        "Celsius"       : temp * 5/4,
        "Fahrenheit"    : (temp * 9/4) + 32,
        "Kelvin"        : (temp * 5/4) + 273.15,
        "Rankine"       : (temp * 9/4) + 491.67
    }

# FUNCION PARA UTILIZAR LA CONVERTIDORA
def convertir_temperatura(temperatura, unidad_inicial, unidad_final):
    if unidad_inicial == unidad_final:      # Agrego esto para que en caso de que la unidad elegida sea la misma no haga la conversión.
        return temperatura
    
    if unidad_inicial == "Celsius":
        conversiones = celsius_a_otras(temperatura)
    elif unidad_inicial == "Fahrenheit":
        conversiones = fahrenheit_a_otras(temperatura)
    elif unidad_inicial == "Kelvin":
        conversiones = kelvin_a_otras(temperatura)
    elif unidad_inicial == "Rankine":
        conversiones = rankine_a_otras(temperatura)
    elif unidad_inicial == "Reaumur":
        conversiones = reaumur_a_otras(temperatura)

    return conversiones[unidad_final]

#LLAMO A LA FUNCION DE SELECCION DE UNIDADES DE TEMPERATURA. CADA UNA CON EL MENSAJE DIFERENTE
mensaje_inicial      = "Seleccione la unidad de temperatura inicial => 1: Celsius (C), 2: Fahrenheit (F), 3: Kelvin (K), 4: Rankine, 5: Reaumur: "
mensaje_transformar1 = "Seleccione la unidad de temperatura a convertir => 1: Celsius (C), 2: Fahrenheit (F), 3: Kelvin (K), 4: Rankine, 5: Reaumur: "
mensaje_transformar2 = "Seleccione otra unidad de temperatura a la que quiera transformar => 1: Celsius (C), 2: Fahrenheit (F), 3: Kelvin (K), 4: Rankine, 5: Reaumur: "

# SOLICITO AL USUARIO LOS DATOS PARA CONVERTIR
temperatura = ingresar_grados()
unidad_i  = seleccionar_unidad(mensaje_inicial)       # Primera vez
unidad_f1 = seleccionar_unidad(mensaje_transformar1)  # Segunda vez
unidad_f2 = seleccionar_unidad(mensaje_transformar2)  # Tercera vez

# EJECUTO LA CONVERSIÓN
conversion1 = convertir_temperatura(temperatura, unidad_i, unidad_f1)
conversion2 = convertir_temperatura(temperatura, unidad_i, unidad_f2)

# IMPRIMO LA CONVERSIÓN
print(f"Convertidor de temperaturas \nSu selección de {temperatura}º {unidad_i} equivale a:\n- {conversion1}º {unidad_f1}\n- {conversion2}º {unidad_f2}")
````

### Exercici 3

Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

#####  CONTADOR DE PALABRAS

#####  INSTRUCCIONES

1. **Ingrese el texto:**
   - Introduzca el texto del cual desea contar las palabras. Puede incluir cualquier tipo de texto con o sin signos de puntuación.

2. **Eliminación de signos de puntuación:**
   - La función eliminará automáticamente los siguientes signos de puntuación del texto:
     ```plaintext
     !()-[]{};:",<>./¡¿?@#$%^&*_~
     ```

3. **Conversión a minúsculas y división en palabras:**
   - El texto será convertido a minúsculas para evitar diferencias entre palabras que solo difieren en mayúsculas/minúsculas.
   - Luego, el texto se dividirá en una lista de palabras.

4. **Cálculo de la cantidad total de palabras:**
   - Se contará el número total de palabras en el texto.

5. **Conteo de la frecuencia de cada palabra:**
   - Se contará cuántas veces aparece cada palabra en el texto.
   - La frecuencia de cada palabra se almacenará en un diccionario.

6. **Ordenación del diccionario de palabras:**
   - El diccionario de palabras y sus frecuencias se ordenará alfabéticamente.

7. **Resultados:**
   - La función devolverá:
     - La lista de palabras en el texto.
     - El número total de palabras.
     - El diccionario ordenado de palabras y sus frecuencias.

8. **Impresión de resultados:**
   - Se imprimirá el texto original ingresado.
   - Se imprimirá la cantidad total de palabras.
   - Se imprimirá el conteo de palabras, ordenadas alfabéticamente, junto con la frecuencia de cada palabra.



````python
# CREO LA FUNCIÓN 
def contar_palabras(texto):
    # Variable que contiene los signos de puntuación a eliminar del texto
    signos_puntuacion = '''!¡()-[]{};:'",<>./¿?@#$%^&*_~'''
    
    # Elimino los signos de puntuación del texto
    for signo in signos_puntuacion:
        texto = texto.replace(signo, '')  # Reemplazo cada signo de puntuación por una cadena vacía

    # Convierto el texto a minúsculas y lo divido en una lista de palabras
    palabras = texto.lower().split()
    len_palabras = len(palabras)  # Cuento el número total de palabras

    # Contar la frecuencia de cada palabra
    conteo_palabras = {}
    for palabra in palabras:
        if palabra in conteo_palabras:
            conteo_palabras[palabra] += 1  # Incremento la cuenta si la palabra ya está en el diccionario
        else:
            conteo_palabras[palabra] = 1   # Agrego la palabra al diccionario con una cuenta inicial de 1

    # Ordenar el diccionario por orden alfabético
    conteo_palabras_ordenado = dict(sorted(conteo_palabras.items()))

    # Devuelvo la lista de palabras, el número total de palabras y el diccionario ordenado de conteo de palabras
    return palabras, len_palabras, conteo_palabras_ordenado 

# Solicito un texto de ejemplo al usuario
texto = input("Ingrese el texto: ")

# Llamo a la función contar_palabras con el texto de ejemplo
palabras, len_palabras, conteo_palabras = contar_palabras(texto)

# Imprimir los resultados de manera clara
print("Texto original:", texto)
print("Cantidad total de palabras:", len_palabras)
print("\nConteo de palabras (ordenadas alfabéticamente):")
for palabra, conteo in conteo_palabras.items():
    print(f"{palabra}: {conteo}")
````

#### Ejercicio 4
Diccionario inverso.

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiados las claves y valores. Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia.

----

#####  FUNCIÓN PARA INVERTIR CLAVES Y VALORES DE UN DICCIONARIO

#####  INSTRUCCIONES

1. **Definición de la función `inverso`:**
   - La función `inverso` toma un parámetro llamado `dictionary`, que es el diccionario que se desea invertir.

2. **Verificación de la unicidad de valores:**
   - Se verifica que los valores en el diccionario sean únicos.
   - Para esto, se utiliza la función `set` que elimina los duplicados. Si la longitud del diccionario original no es igual a la longitud del conjunto de valores, significa que hay valores repetidos.
   - En caso de valores duplicados, se imprime el mensaje "Error: multiple keys for one value" y la función devuelve `None`, indicando que no es posible invertir el diccionario.

3. **Inversión de claves y valores:**
   - Si los valores son únicos, se crea un nuevo diccionario donde las claves y valores del diccionario original se intercambian.
   - Este nuevo diccionario se devuelve como resultado.

4. **Resultado si los valores son duplicados:**
   - Si los valores se duplican, tal como lo indica el enunciado, entoces el resultado que se devuelve es el mensaje de error.


````python
# DEFINICIÓN DE LA FUNCIÓN INVERSO: DEFINO FUNCIÓN LLAMADA INVERSO QUE TOMA UN PARÁMETRO 'DICTIONARY' 
def inverso(dictionary):
    # VERIFICO DE UNICIDAD DE VALORES: 
    # UITILIZO LA FUNCIÓN 'SET', YA QUE ASÍ ME CREA UN SET DE DATOS QUE SERÁN UNICOS.
    if len(dictionary.values()) != len(set(dictionary.values())): 
        # SI EL LEN DE 'DICTIONARY' ES DISTINTO A LEN(SET) SIGNIFICA QUE HAY ELEMENTROS REPETIDOS.
        print("Error: multiple keys for one value")
        return None # la funcion devuelve 'none' para indicar que no puede invertirlos
    
    # INVIERTO LAS CLAVES Y VALORES DEL DICCIONARIO: -
    # CREO UN NUEVO DICCIONARIO DONDE CADA CLAVE K Y VALOR V EN EL DICCIONARIO INTERCAMBIAN DE LUGAR.
    orden_inverso = {v: k for k, v in dictionary.items()} 
    # FINALMENTE TENGO LA FUNCIÓN TERMINADA
    return orden_inverso


# PRUEBAS

# PRUEBA 1: DICCIONARIO CON VALORES ÚNICOS

dictionary = {"a": 1, "b": 2, "c": 3}           # Defino un diccionario con valores únicos.
print("Diccionario original Prueba 1:", dictionary)
resultado = inverso(dictionary)                 # Llamo a la función inverso y almaceno el resultado.
if resultado is not None:                       # Verifico si error de valores duplicados).
   print("Diccionario inverso  Prueba 1:", resultado)   # Imprimo el diccionario invertido.

print("----")
# PRUEBA 2:  DEFINO UN DICCIONARIO CON VALORES DUPLICADOS.
dictionary = {"x": "apple", "y": "banana", "z": "banana"}    
print("Diccionario original Prueba 2:", dictionary) 
resultado = inverso(dictionary)                 # Llamo a la función inverso y almaceno el resultado.
if resultado is not None:                       # Verifico si el resultado no es None                                       
    print("Diccionario inverso Prueba 2:", resultado)  # Imprimo el diccionario invertido.
````

### Exercici 1
#### Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.

---

#### Instrucciones de Uso

#### Generar el diccionario aleatorio:

1. Se define una función `generate_random_dict_and_group` que genera un diccionario aleatorio con letras minúsculas como claves y números enteros del 1 al 9 como valores.
2. También agrupa las letras por sus valores en un `defaultdict` de listas.

#### Separar y agrupar el diccionario:

1. Se define una función `separar_invertir` que toma el diccionario aleatorio, agrupa las letras por sus valores y devuelve este agrupamiento en un `defaultdict` de listas.

#### Ejecutar el código:

1. Se genera un diccionario aleatorio de tamaño 5.
2. Se imprime el diccionario generado.
3. Se agrupa el diccionario por valores y se invierten las claves y valores.
4. Se imprime el diccionario invertido.

#### Imprimir resultados:

1. El diccionario generado se muestra con las claves originales y los valores asignados.
2. El diccionario invertido se muestra con los valores originales como claves y las letras correspondientes como valores.
----


````python
import random
import string
from collections import defaultdict

# Función para generar un diccionario aleatorio y agrupar sus valores
def generate_random_dict_and_group(size):
    random_dict = {}
    
    # Generar el diccionario aleatorio
    for _ in range(size):
        letra = random.choice(string.ascii_lowercase)
        valor_numerico = random.randint(1, 9)
        random_dict[letra] = valor_numerico 
    
    # Agrupar los valores en un defaultdict de listas
    grouped_dict = defaultdict(list)
    for key, value in random_dict.items():
        grouped_dict[value].append(key) 
    
    return random_dict, grouped_dict

# Tamaño del diccionario aleatorio
size = 5

# Generar el diccionario aleatorio y el diccionario agrupado
random_dict, grouped_dict = generate_random_dict_and_group(size)

# Imprimir el diccionario generado
print("Diccionario generado:\n", random_dict)

# Función para separar y agrupar el diccionario, luego invertir sus claves y valores
def separar_invertir(random_dict):
    grouped_dict = defaultdict(list)
    
    # Agrupar los valores en un defaultdict de listas
    for key, value in random_dict.items():
        grouped_dict[value].append(key)
    
    # Convertir listas de un solo elemento a cadenas
    for key in grouped_dict:
        if len(grouped_dict[key]) == 1:
            grouped_dict[key] = grouped_dict[key][0]
    
    return grouped_dict

# Separar e invertir el diccionario
sep = dict(separar_invertir(random_dict))

# Imprimir el diccionario inverso:
print("Diccionario inverso: \n", sep)

# Imprimir el resultado
print("Diccionario separado e invertido:")
for key, value in sep.items():
    print(f"{key}: {value}")


````

### Exercici 2
Conversió de tipus de dades

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

----

````python
# LISTA PROPORCIONADA POR EL CLIENTE
lst = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

# FUNCIÓN PARA QUE CADA ELEMENTO SEA UN ELEMENTO DE LA LISTA.
# GENERADOR QUE PRODUCE CADA ELEMENTO DE LA LISTA APLANADA.
def aplanar(lst):
    for item in lst:
        if isinstance(item, (list, tuple)):
            yield from aplanar(item)
        else:
            yield item

# OBTENER LA LISTA APLANADA
lista_aplanada = list(aplanar(lst))

# INICIALIZAR LISTAS PARA ENTEROS Y ELEMENTOS NO CONVERTIDOS
convertidos = []
no_convertidos = []

# ITERAR SOBRE CADA ELEMENTO DE LA LISTA APLANADA
for item in lista_aplanada:
    try:
        # Convertir el elemento a float
        numero = float(item)
        convertidos.append(numero)  # Agregar a la lista de convertidos
    except ValueError:
        no_convertidos.append(item)  # Si la conversión falla, agregar a la lista de no convertidos
        
# IMPRIMIR RESULTADOS
print("lista inicial:           ", lst)
print("Lista aplanada:          ", lista_aplanada)
print("Números convertidos:     ", convertidos)
print("Elementos no convertidos:", no_convertidos)
````

````python
# Función para habilitar el contenido del archivo
def habilitar_texto(txt_cliente):
    with open(txt_cliente, "r", encoding="utf-8") as archivo:
        return archivo.read()

# Función para contar y ordenar las palabras
def contar_palabras(texto):
    # Definición del alfabeto 
    alfabeto = 'abcdefghijklmnopqrstuvwxyz'
    
    # Definir un diccionario vacío para almacenar las palabras por letra inicial
    diccionario_paraules = {letra: {} for letra in alfabeto}
    
    # Eliminar signos de puntuación y dividir el texto en palabras
    signos_puntuacion = '''!()-[]{};:'",<>./¡¿?@#$%^&*_~'''
    for signo in signos_puntuacion:
        texto = texto.replace(signo, '')
    
    palabras = sorted(texto.lower().split())
    
    # Contar la frecuencia de cada palabra y agruparlas por letra inicial
    for palabra in palabras:
        primera_letra = palabra[0]
        if primera_letra in diccionario_paraules:
            if palabra in diccionario_paraules[primera_letra]:
                diccionario_paraules[primera_letra][palabra] += 1
            else:
                diccionario_paraules[primera_letra][palabra] = 1
    
    # Ordenar las palabras dentro de cada letra inicial
    diccionario_ordenat = {}
    for letra, paraules_dict in diccionario_paraules.items():
        if paraules_dict:
            diccionario_ordenat[letra] = dict(sorted(paraules_dict.items()))
    
    return diccionario_ordenat

# Nombre del archivo cliente
txt_cliente = "tu_me_quieres_blanca.txt"

# Habilitar el texto del archivo
texto = habilitar_texto(txt_cliente)

# Contar y ordenar las palabras
diccionario_ordenat = contar_palabras(texto)

# Imprimir el diccionario ordenado de manera detallada
print("Diccionario ordenado:")
for letra, palabras_dict in diccionario_ordenat.items():
    print(f"{letra}:")
    for palabra, frecuencia in palabras_dict.items():
        print(f"  {palabra}: {frecuencia}")
````
